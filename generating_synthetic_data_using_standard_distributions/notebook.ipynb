# Generating Synthetic Data using Standard distributions

Finance Tech company located in Canda, trying to determine the risk factor for customers. The synthetic dataset to be created should have the following attributes:

### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
import seaborn as sns
```

### Age

Canada working-age population range 15-64.

Mean = (64+15)/2 = 39.5

Standard deviation = (64-15)/4 = 12.25

[Canada working-age population](https://www150.statcan.gc.ca/n1/daily-quotidien/220427/mc-a003-eng.htm)

```python
# Define the parameters for the normal distribution
mean_age = 39.5
std_dev_age = 12.25
num_samples = 1500  # Number of synthetic data points to generate

# Generate synthetic ages
synthetic_ages = np.random.normal(loc=mean_age, scale=std_dev_age, size=num_samples)

# Round the ages to the nearest integer
synthetic_ages = np.round(synthetic_ages)

# Filter out ages outside the range of 15 to 64 years
synthetic_ages = synthetic_ages[(synthetic_ages >= 15) & (synthetic_ages <= 64)]

# Cap at 1000 samples
synthetic_ages = synthetic_ages[:1000]

# Plot the distribution of the synthetic ages to visualize
plt.hist(synthetic_ages, bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Synthetic Ages (15 to 64 years)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Display first 10 synthetic ages
print(len(synthetic_ages))
```

### Gender

Let's have a close difference between the genders

Male = 55%

Female = 45%


```python
# Define the gender distribution
gender_distribution = ['Male', 'Female']
gender_probabilities = [0.55, 0.45]
num_samples = 1000  # Number of synthetic data points to generate

# Generate synthetic genders
synthetic_genders = np.random.choice(gender_distribution, size=num_samples, p=gender_probabilities)

# Plot the distribution of the synthetic genders to visualize
plt.hist(synthetic_genders, bins=np.arange(len(gender_distribution)+1)-0.5, edgecolor='k', alpha=0.7)
plt.xticks(np.arange(len(gender_distribution)), gender_distribution)
plt.title('Distribution of Synthetic Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()

# Display first 10 synthetic genders
print(synthetic_genders[:10])
```

### Job Title, Years at Current Job & Monthly Income

#### Job title & Monthly Income
[Common jobs in Canada](https://ca.indeed.com/career-advice/finding-a-job/common-jobs)

[Highest paid jobs in Canada](https://www.simplilearn.com/highest-paying-jobs-in-canada-article)

#### Years at current job

Years range: 1 to 5 years

https://www.statista.com/statistics/438998/employment-level-in-canada-by-job-tenure/

```python
# Define the job titles and their associated monthly incomes
job_titles = [
    'Cashier', 'Food preparation worker', 'Janitor', 'Bartender', 'Server',
    'Business Intelligence Analyst', 'Cloud Architect', 'Cloud Systems Engineer',
    'IT Project Manager', 'Data Scientist'
]
monthly_incomes = {
    'Cashier': 3574,
    'Food preparation worker': 3794,
    'Janitor': 3566,
    'Bartender': 3299,
    'Server': 3593,
    'Business Intelligence Analyst': 6917,
    'Cloud Architect': 11416,
    'Cloud Systems Engineer': 8083,
    'IT Project Manager': 8833,
    'Data Scientist': 8417
}

num_samples = 1000  # Number of synthetic data points to generate

# Generate synthetic job titles
synthetic_job_titles = np.random.choice(job_titles, size=num_samples)

# Generate synthetic monthly incomes based on job titles
synthetic_monthly_incomes = [monthly_incomes[job_title] for job_title in synthetic_job_titles]

# Generate synthetic years at current job
synthetic_years_at_current_job = np.random.randint(1, 6, size=num_samples)

# Visualize the distributions
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot job titles
axs[0].hist(synthetic_job_titles, bins=np.arange(len(job_titles) + 1) - 0.5, edgecolor='k', alpha=0.7)
axs[0].set_xticks(np.arange(len(job_titles)))
axs[0].set_xticklabels(job_titles, rotation=90)
axs[0].set_title('Distribution of Synthetic Job Titles')
axs[0].set_xlabel('Job Title')
axs[0].set_ylabel('Frequency')

# Plot monthly incomes
axs[1].hist(synthetic_monthly_incomes, bins=30, edgecolor='k', alpha=0.7)
axs[1].set_title('Distribution of Synthetic Monthly Incomes')
axs[1].set_xlabel('Monthly Income')
axs[1].set_ylabel('Frequency')

# Plot years at current job
axs[2].hist(synthetic_years_at_current_job, bins=np.arange(1, 7) - 0.5, edgecolor='k', alpha=0.7)
axs[2].set_xticks(np.arange(1, 6))
axs[2].set_title('Distribution of Synthetic Years at Current Job')
axs[2].set_xlabel('Years at Current Job')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Display first 10 synthetic data points
for i in range(10):
    print(f"Job Title: {synthetic_job_titles[i]}, Monthly Income: ${synthetic_monthly_incomes[i]}, Years at Current Job: {synthetic_years_at_current_job[i]}")
```

### Dependency on Credit

[Statistics for Cash and Credit Card Use for Payments in 2024](https://www.clearlypayments.com/blog/statistics-for-cash-and-credit-card-use-for-payments-in-2024/)

```python
# Define the dependence on credit distribution
dependence_on_credit_distribution = ['Medium', 'High', 'Low']
dependence_on_credit_probabilities = [0.70, 0.15, 0.15]
num_samples = 1000  # Number of synthetic data points to generate

# Generate synthetic dependence on credit
synthetic_dependence_on_credit = np.random.choice(dependence_on_credit_distribution, size=num_samples, p=dependence_on_credit_probabilities)

# Plot the distribution of the synthetic dependence on credit to visualize
plt.hist(synthetic_dependence_on_credit, bins=np.arange(len(dependence_on_credit_distribution) + 1) - 0.5, edgecolor='k', alpha=0.7)
plt.xticks(np.arange(len(dependence_on_credit_distribution)), dependence_on_credit_distribution)
plt.title('Distribution of Synthetic Dependence on Credit')
plt.xlabel('Dependence on Credit')
plt.ylabel('Frequency')
plt.show()

# Display first 10 synthetic dependence on credit
print(synthetic_dependence_on_credit[:10])
```

### Risk Appetite

https://www.jstor.org/stable/23859528#:~:text=Low%20income%20groups%20still%20use,Graduate%20School%20of%20Management%201989).&text=ment%20on%20their%20credit%20cards,Boston%20Company%20Economic%20Advisor%20Inc.

```python
# Define a function to calculate the risk appetite based on other attributes
def assign_risk_appetite(marital_status, employment_status, monthly_income, dependence_on_credit):
    if marital_status == 'Single' and employment_status in ['Employed', 'Self-Employed'] and monthly_income > 5000 and dependence_on_credit == 'High':
        return 'High'
    elif marital_status == 'Married' and employment_status in ['Employed', 'Self-Employed'] and monthly_income > 3000 and dependence_on_credit in ['Medium', 'High']:
        return 'Medium'
    else:
        return 'Low'

# Generate synthetic data for other attributes (example values)
num_samples = 1000
# Assign risk appetite based on the synthetic attributes
synthetic_risk_appetites = [
    assign_risk_appetite(
        marital_status, employment_status, monthly_income, dependence_on_credit
    )
    for marital_status, employment_status, monthly_income, dependence_on_credit
    in zip(
        synthetic_marital_statuses, synthetic_employment_statuses, synthetic_monthly_incomes, synthetic_dependence_on_credit
    )
]

```

### Customer name

```python
# Predefined lists of first names and last names
first_names = [
    'James', 'John', 'Robert', 'Michael', 'William', 'David', 'Richard', 'Joseph', 'Thomas', 'Charles',
    'Mary', 'Patricia', 'Jennifer', 'Linda', 'Elizabeth', 'Barbara', 'Susan', 'Jessica', 'Sarah', 'Karen'
]
last_names = [
    'Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Rodriguez', 'Martinez',
    'Hernandez', 'Lopez', 'Gonzalez', 'Wilson', 'Anderson', 'Thomas', 'Taylor', 'Moore', 'Jackson', 'Martin'
]

# Function to generate random customer names
def generate_random_customer_names(num_samples):
    return [f"{random.choice(first_names)} {random.choice(last_names)}" for _ in range(num_samples)]

# Number of synthetic data points to generate
num_samples = 1000

# Generate random customer names
synthetic_customer_names = generate_random_customer_names(num_samples)

```

## Dataset  - from Standard Distribution

```python
# Convert lists to pandas Series
synthetic_customer_names = pd.Series(synthetic_customer_names)
synthetic_ages = pd.Series(synthetic_ages)
synthetic_genders = pd.Series(synthetic_genders)
synthetic_marital_statuses = pd.Series(synthetic_marital_statuses)
synthetic_employment_statuses = pd.Series(synthetic_employment_statuses)
synthetic_job_titles = pd.Series(synthetic_job_titles)
synthetic_years_at_current_job = pd.Series(synthetic_years_at_current_job)
synthetic_monthly_incomes = pd.Series(synthetic_monthly_incomes)
synthetic_dependence_on_credit = pd.Series(synthetic_dependence_on_credit)
synthetic_risk_appetites = pd.Series(synthetic_risk_appetites)

column_data = {
    'Customer name': synthetic_customer_names,
    'Age': synthetic_ages,
    'Gender': synthetic_genders,
    'Marital Status': synthetic_marital_statuses,
    'Employment Status': synthetic_employment_statuses,
    'Job Title': synthetic_job_titles,
    'Years at Current Job': synthetic_years_at_current_job,
    'Monthly Income': synthetic_monthly_incomes,
    'Dependence on Credit': synthetic_dependence_on_credit,
    'Risk Appetite': synthetic_risk_appetites
}


fin_tech = pd.DataFrame(column_data)
fin_tech.head(10)
```
